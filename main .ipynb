```python
#Bài 1: Đọc và kiểm tra dữ liệu
import pandas as pd

# 1. Đọc file vừa tạo
df = pd.read_csv("student_performance_dirty.csv")

# 2. Hiển thị 5 dòng đầu tiên để nhìn tổng quan
print("--- 5 DÒNG ĐẦU TIÊN ---")
print(df.head())

# 3. Kiểm tra thông tin các cột (số lượng, kiểu dữ liệu)
print("\n--- THÔNG TIN CẤU TRÚC DỮ LIỆU ---")
print(df.info())

# 4. Đếm số lượng giá trị bị thiếu (NaN) ở từng cột
print("\n--- THỐNG KÊ DỮ LIỆU BỊ THIẾU ---")
print(df.isna().sum())
```

```python
#Bài 2: Chuẩn hóa tên lớp và giới tính

# 1. Chuẩn hóa tên lớp (class_name)
df["class_name"] = (df["class_name"]
                    .str.replace("-", "", regex=False)        # Xóa dấu gạch ngang (CNTT-3 -> CNTT3)
                    .str.replace(r"\s+", " ", regex=True)     # Thu gọn nhiều khoảng trắng thành 1 khoảng trắng
                    .str.strip()                              # Xóa khoảng trắng ở 2 đầu
                    .str.upper()                              # Viết hoa toàn bộ
                    .replace({"CNTT1": "CNTT 1", "CNTT2": "CNTT 2", "CNTT3": "CNTT 3"})) # Đưa về chuẩn cuối cùng

# 2. Chuẩn hóa giới tính (gender)
df["gender"] = (df["gender"]
                .astype("string")       # Ép kiểu về chuỗi
                .str.strip()            # Xóa khoảng trắng 2 đầu
                .str.lower()            # Đưa về chữ thường để dễ so sánh (nam, nữ, nu)
                .replace({"nam": "Nam", "nữ": "Nữ", "nu": "Nữ"})) # Ánh xạ về chuẩn có dấu viết hoa

print("Dữ liệu sau khi chuẩn hóa Lớp và Giới tính:")
print(df[["student_id", "class_name", "gender"]].head(7))
```

```python
#Bài 3: Lấp đầy dữ liệu thiếu

# 1. Cột phân loại (Giới tính): Lấp bằng chuỗi mặc định
df["gender"] = df["gender"].fillna("Không rõ")

# 2. Cột số liệu (Tỷ lệ chuyên cần): Lấp bằng Trung vị (median)
# Trung vị an toàn hơn Trung bình (mean) vì nó không bị ảnh hưởng bởi các số quá lớn hoặc quá nhỏ
df["attendance_rate"] = df["attendance_rate"].fillna(df["attendance_rate"].median())

# 3. Cột thông tin liên hệ (Số điện thoại): Lấp bằng chuỗi thông báo
df["phone"] = df["phone"].fillna("Chưa cập nhật")

print("Kiểm tra lại xem còn dữ liệu thiếu ở 3 cột này không:")
print(df[["gender", "attendance_rate", "phone"]].isna().sum())
```

```python
#Bài 4: Loại bỏ bản ghi trùng lặp

# 1. Kiểm tra xem có bao nhiêu dòng bị trùng lặp hoàn toàn (giống hệt nhau ở mọi cột)
so_dong_trung = df.duplicated().sum()
print(f"Số dòng trùng lặp toàn bộ: {so_dong_trung}")

# 2. Kiểm tra trùng lặp dựa trên mã sinh viên (student_id)
# Đôi khi thông tin bị lệch một chút nhưng cùng 1 mã SV thì vẫn là trùng
so_sv_trung = df.duplicated(subset=["student_id"]).sum()
print(f"Số mã sinh viên bị trùng: {so_sv_trung}")

# 3. Xóa các dòng trùng lặp theo mã sinh viên
# keep="first" nghĩa là giữ lại bản ghi xuất hiện đầu tiên, xóa các bản sao bên dưới
df = df.drop_duplicates(subset=["student_id"], keep="first")

print("Dữ liệu sau khi xóa trùng lặp (Đã xóa SV005 bản sao):")
print(df[["student_id", "full_name"]].head(6))
```

```python
#Bài 5: Phát hiện điểm số không hợp lệ

# 1. Lọc ra các điểm số KHÔNG nằm trong khoảng từ 0 đến 10
# Hàm between(0, 10) kiểm tra xem điểm có nằm trong khoảng 0-10 không.
# Dấu ~ (ngã) ở trước mang ý nghĩa là phủ định (NOT), tức là tìm những điểm NẰM NGOÀI khoảng đó.
invalid_score = ~df["score_python"].between(0, 10)

print("Các sinh viên có điểm Python bị lỗi:")
print(df.loc[invalid_score, ["student_id", "score_python"]])

# 2. Xử lý: Biến các điểm lỗi này thành giá trị thiếu (NaN)
df.loc[invalid_score, "score_python"] = np.nan

# 3. Lấp đầy lại các điểm lỗi (đã thành NaN) bằng điểm Trung vị của cả lớp
df["score_python"] = df["score_python"].fillna(df["score_python"].median())

print("\nĐiểm của sinh viên bị lỗi (SV005) sau khi được lấp bằng trung vị:")
print(df[df["student_id"] == "SV005"][["student_id", "score_python"]])
```

```python
#Bài 6: Chuẩn hóa chuỗi và Regex

# 1. Chuẩn hóa họ tên: Cắt khoảng trắng thừa và viết hoa chữ cái đầu
df["full_name"] = (df["full_name"]
                   .str.replace(r"\s+", " ", regex=True) # Đưa nhiều khoảng trắng về 1
                   .str.strip()                          # Cắt râu ria 2 đầu
                   .str.title())                         # Viết hoa chữ cái đầu (VD: nguyen van a -> Nguyen Van A)

# 2. Kiểm tra email bằng Regex (Biểu thức chính quy)
# Mẫu regex này yêu cầu: [chữ/số] + @ + [chữ/số] + . + [chữ]
email_mask = df["email"].str.contains(r"^[\w\.-]+@[\w\.-]+\.\w+$", regex=True, na=False)

print("Các sinh viên có email sai định dạng (thiếu @ hoặc dấu chấm):")
print(df.loc[~email_mask, ["student_id", "email"]])

# 3. Chuẩn hóa số điện thoại: Chỉ giữ lại các chữ số
# \D trong Regex nghĩa là "Tất cả những gì KHÔNG PHẢI là số"
df["phone"] = df["phone"].astype("string").str.replace(r"\D", "", regex=True)

print("\nDữ liệu sau khi dọn dẹp chuỗi:")
print(df[["student_id", "full_name", "email", "phone"]].head())
```

```python
#Bài 7: Chuyển kiểu ngày tháng

# errors="coerce" sẽ biến những ngày tháng nhập sai hoàn toàn (như "ngày 32 tháng 13") thành NaT (Not a Time)
df["birth_date"] = pd.to_datetime(df["birth_date"], errors="coerce", dayfirst=True)

so_ngay_loi = df["birth_date"].isna().sum()
print(f"Số lượng ngày sinh không hợp lệ hoặc bị thiếu sau khi chuyển đổi: {so_ngay_loi}")
```

```python
#Bài 8: Phân nhóm dữ liệu (Binning)

# Xác định các mốc điểm cắt
bins = [0, 5, 6.5, 8, 10]
# Đặt tên cho từng khúc
labels = ["Yếu", "Trung bình", "Khá", "Giỏi"]

# Dùng hàm pd.cut() để cắt điểm vào các rổ tương ứng
df["level"] = pd.cut(df["score_python"], bins=bins, labels=labels, include_lowest=True)

print("Kết quả xếp loại học lực:")
print(df[["student_id", "score_python", "level"]].head(7))
```

```python
#Bài 9: Phát hiện outlier bằng IQR

# 1. Tính các mốc Tứ phân vị
q1 = df["attendance_rate"].quantile(0.25) # Mốc 25%
q3 = df["attendance_rate"].quantile(0.75) # Mốc 75%
iqr = q3 - q1                             # Khoảng tứ phân vị

# 2. Tính giới hạn an toàn (trên và dưới)
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

# 3. Lọc ra những sinh viên nằm NGOÀI giới hạn an toàn
outlier_df = df[(df["attendance_rate"] < lower) | (df["attendance_rate"] > upper)]

print(f"Khoảng an toàn của tỷ lệ chuyên cần: Từ {lower:.2f} đến {upper:.2f}")
print("Danh sách các bản ghi nghi ngờ là Outlier (cần giảng viên kiểm tra):")
print(outlier_df[["student_id", "attendance_rate"]])
```

```python
#Bài 10: Xuất file và nhận xét

# Xuất file CSV (nhớ dùng utf-8-sig để không bị lỗi font tiếng Việt khi mở bằng Excel)
df.to_csv("student_performance_clean.csv", index=False, encoding="utf-8-sig")

print("✅ Đã lưu thành công file: 'student_performance_clean.csv'")

# Yêu cầu viết nhận xét ngắn [cite: 150]
nhan_xet = """
Nhận xét các bước xử lý quan trọng:
1. Xử lý Missing Data: Đã lấp đầy dữ liệu thiếu ở giới tính, tỷ lệ chuyên cần (dùng trung vị) và số điện thoại.
2. Xóa Trùng Lặp: Đã loại bỏ thành công bản ghi copy thừa của sinh viên SV005.
3. Chuẩn hóa & Regex: Làm sạch khoảng trắng họ tên, chuẩn hóa định dạng Lớp/Giới tính, và dùng Regex lọc thành công email sai định dạng, gọt sạch chữ cái lẫn trong số điện thoại.
4. Xử lý Logic & Ngoại lệ: Cắt bỏ điểm Python vô lý (11.0) và dùng thuật toán IQR tóm được các tỷ lệ điểm danh bất thường.
"""
print(nhan_xet)
```
